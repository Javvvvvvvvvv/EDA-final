```python
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
```

```python
class ColumnSelector(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X[self.columns]
```

```python

# Create a pipeline
pipeline = Pipeline([
    # Handle missing values with median imputation
    ('select_columns', ColumnSelector(columns=['age', 'fare'])),
    ('impute_missing', SimpleImputer(strategy='median')),  
    ('scale_data', StandardScaler()),
    ('pca', PCA(n_components=2)),
])
```

```python
# Fit and transform the data using the pipeline
principal_components = pipeline.fit_transform(titanic)

# Create a DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['age', 'fare'])

# Plot the principal components
plt.figure(figsize=(10, 6))
plt.scatter(principal_df['age'], principal_df['fare'], c=titanic['survived'], cmap='viridis')
plt.xlabel('Passenger Age')
plt.ylabel('Passenger Fare')
plt.title('PCA on Titanic Dataset')
plt.colorbar().set_label('Survived')
plt.show()
```
